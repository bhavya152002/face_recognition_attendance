```python
#Defining properties of the second window
def Section(sec):
    print(f"The section is {sec}")


def camera(x):

    Recognize.recognize_attendence()
    key = input("Enter any key to return main menu")

def result(x):
    if x == 1:
        top = Toplevel()
    lbl = Label(top, text='RESULT')
    with open("python/projects/Face-Recognition-Attendance-System/FRAS/StudentDetails/StudentDetails.csv", newline = "") as file:
        reader = csv.reader(file)

        # r and c tell us where to grid the labels
        r = 0
        for col in reader:
            c = 0
            for row in col:
                # i've added some styling
                label = tkinter.Label(top, width = 10, height = 2, \
                                      text = row, relief = tkinter.RIDGE)
                label.grid(row = r, column = c)
                c += 1
            r += 1
    

        

#Defining properties of the first window

def bttn(root,text,bcolor,fcolor,font,cmd):

    def on_enter(e):
        myButton['background']=bcolor
        myButton['foregrounf']=fcolor

    def on_leave(e):
        myButton['background']=fcolor
        myButton['foregrounf']=bcolor



    myButton=Button(root,height=2,width=400,text=text,
                    fg=bcolor,
                    bg=fcolor,
                    border=0,
                    activeforeground=fcolor,
                    activebackground=bcolor,
                    command=cmd)

    myButton.bind("<Enter>", on_enter)
    myButton.bind("<Leave>", on_leave)
    myButton.pack()

def cmd():
    top = Toplevel()
    top.geometry('400x400')
    lbl = Label(top, text='Select Section').pack()
    but = Button(top, text='A', command=lambda: Section('A')).pack(pady=10)
    but1 = Button(top, text='B', command=lambda: Section('B')).pack(pady=10)
    cam = Button(top, text='CAMERA', command=lambda: camera(1)).pack(pady=10)
    resultbut = Button(top, text='Result', command=lambda: result(1)).pack(pady=10)
```
